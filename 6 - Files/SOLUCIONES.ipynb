### Escribir un código que almacene en un fichero autores y libros y que posteriormente permita realizar consultas.

Crear una función **Creacion(file, n)** que le pida al usuario n veces el nombre de un autor y un libro y lo añada al archivo cuyo nombre se le pasa como parametro a la funcion. 

Una vez creado el archivo, debe escribir las siguientes funciones:

* **Display(f)** para visualización del contenido del archivo mostrando los libros y autores contenidos.
* **Search(f,nombre)** para buscar todos los libros escritos por un autor. Muestre los títulos y el número de libros escritos por ese autor. Si el nombre del autor no existe en el archivo, muestra un mensaje de error.
* **Menu()** para mostrar las 3 opciones que le daremos al usuario: 
    1. Mostrar el contenido del fichero
    2. Buscar libros de un autor
    3. Salir.

### **Solución:**

```python
def Creacion(f,n):
    for i in range(n):
        autor=input("Introduzca el nombre del autor {0:2d}: ".format(i+1))
        titulo=input("Introduzca el libro de {0:2s}: ".format(autor))
        f.write(autor+"\n"+titulo+"\n")
    return


def Menu():
    print("\nMENU DE OPCIONES")
    print("================")
    print("1. Mostrar el contenido del fichero")
    print("2. Buscar libros de un autor")
    print("3. Salir")
    opc=int(input("Introduzca la opcion deseada: "))
    while opc<1 or opc>3:
        print("Error, numero invalido")
        opc=int(input("Introduzca la opcion deseada: "))
    return opc


def Display(f):
    f.seek(0)
    datos=f.read()
    print("\nCONTENIDO DEL FICHERO")
    print("------------------------")
    print(datos)
    return


def Search(f,nombre):
    f.seek(0)
    lista=f.read().splitlines()
    while nombre not in lista:
        print("Error, autor no encontrado")
        nombre=input("Introduzca el nombre del autor que desea buscar: ")
    contador=0
    print("LIBROS DE {0:2s}".format(nombre.upper()))
    print("========================")
    for i in range(len(lista)):
        if nombre==lista[i-1]:
            print(lista[i])
            contador=contador+1
    print("{0:2s} tiene {1:2d} libro/s".format(nombre,contador))
    return
```

```python
file=open("books.txt","w+")
num_libros=int(input("Introduzca el numero de libros que desea incluir en el archivo: "))
Creacion(file,num_libros)
resp=Menu()
while resp!=3:
    if resp==1:
        Display(file)
    elif resp==2:
        nombre=input("Introduzca el nombre del autor que desea buscar: ")
        Search(file,nombre)
    resp=Menu()
file.close()
print("Hasta pronto")
```

## Ejercicio 2:

### Escribe un programa que pida al usuario ciudades y temperaturas y genere un fichero "temperaturas.txt" que almacenará una ciudad por línea:

        ciudad1;temperatura1
        ciudad2;temperatura2
        ciudad3;temperatura3

Para ello habrá que crear las siguietnes funcionas:

* **read_temps()**. Esta función pedirá datos al usuario hasta que este escribe la palabra 'exit'. La función devolverá un diccionario con las ciudades como claves y las temperaturas como valores 

        {ciudad1: temperatura1, ciudad2: temperatura2, ...}.

* **write_temps(temps)**. Esta función volcará el diccionario de temperaturas a un fichero CSV
* **read_display()**. Esta función leerá el contenido del fichero **temperaturas.txt** y lo mostrará por pantalla. Después de haber sacado por pantalla todas las ciudades mostrará un mensaje con la ciudad que ha tenido la temperatura máxima:
        
        --- Reading data from user     
        City name: cuenca
        Average temperature for cuenca: 34
        City name [exit to finish]: madrid
        Average temperature for madrid: 33
        City name [exit to finish]: oviedo
        Average temperature for oviedo: 15
        City name [exit to finish]: exit
        
        --- Writing data to file
        
        --- Reading data from file
        City name       Avg Temp
        cuenca            34.0ºC
        madrid            33.0ºC
        oviedo            15.0ºC
        Maximum average temperature:   34.0ºC for cuenca

### **Solución:**

```python
def read_temps():
    temps={}
    name=input("City name: ")
    while name!="exit":
        temp=float(input("Average temperature for "+name+": "))
        temps[name]=temp  #store temp in dictionary temps
        name=input("City name [exit to finish]: ")    
    return temps

def write_temps(temps): #temps is a dictionary with city names and average temperatures
    with open("temps.txt","w") as f:
        for k,v in temps.items():
            f.write("%s;%6.1f\n" % (k,v))
            
def read_display():
    max_temp=-10000  #very low value
    max_city=""
    with open("temps.txt","r") as f:
        line=f.readline()
        print("City name       Avg Temp")
        while (line):
            line_list=line.split(";")
            #print(line_list)   #debugging
            n=line_list[0] #city name
            t=float(line_list[1]) #avg temp
            if t>max_temp:
                max_temp=t
                max_city=n
            print("%-15s %6.1fºC" % (n,t))
            line=f.readline()
    print("Maximum average temperature: %6.1fºC for %s" % (max_temp,max_city))
```

```python
print("--- Reading data from user")
temps=read_temps()
print("--- Writing data to file")
write_temps(temps)
print("--- Reading data from file")
read_display()
```

### **Solución 3.1:**

```python
def Read_TXT(filename):
    res=[] #list of transaction
    with open(filename,"r") as f:
        line=f.readline()
        while line:
            lst=line.split(';')
            #print(lst) #for debugging purposes
            lst2=[lst[0], int(lst[1]), float(lst[2])]
            res.append(lst2)
            line=f.readline()
    return res
```

```python
import pickle
def Write_DAT(filename,trx):
    with open(filename+".dat","wb") as f:
        pickle.dump(trx,f)
```

```python
import pickle
def Read_DAT(filename):
    with open(filename+".dat","rb") as f:
        trx=pickle.load(f)
    print("Product     units  price")
    print("------------------------")
    for tr in trx:
        print("%-13s %3d %6.2f" % (tr[0],tr[1],tr[2]))
```

```python
print("Reading TXT file")
print("================")
trx=Read_TXT("transactions.txt")
print("Writing binary")
print("==============")
Write_DAT("binary",trx)
print("Reading binary")
print("==============")
Read_DAT("binary")
```

### **Solución:**

```python
import pickle

def Leer_y_Grabar_Pedidos(filename, num_pedidos):
    with open(filename,"wb") as f:
        for i in range(num_pedidos):
            cliente=input("Introduzca cliente: ")
            plantas_a=int(input("Introduzca número de plantas de tipo A: "))
            plantas_b=int(input("Introduzca número de plantas de tipo B: "))
            registro=[cliente, plantas_a, plantas_b]
            pickle.dump(registro,f)

def Mostrar_Fichero_Binario(filename, plantas):
    with open(filename,"rb") as f:
        try:
            while True:
                registro=pickle.load(f)
                print("Cliente: {0:10s} Plantas Tipo A: {1} Total A: {2:4.1f} Plantas Tipo B: {3} Total B: {4:4.1f} Coste Total Plantas: {5:5.2f}".format(registro[0], registro[1], registro[1]*plantas[0],registro[2], registro[2]*plantas[1],(registro[1]*plantas[0])+(registro[2]*plantas[1])))
        except:
            pass

def Calcular_Maximos_Clientes(filename, importe_maximo, plantas):
    with open(filename,"rb") as f:
        with open("maximo.csv","w") as maximo:
            try:
                while True:
                    registro=pickle.load(f)
                    r=((registro[1]*plantas[0])+(registro[2]*plantas[1]))
                    if(r>importe_maximo):
                        maximo.write(registro[0]+" "+str(r)+'\n')                
            except:
                pass
            
def Mostrar_Fichero_Texto():
    with open("maximo.csv","r") as maximo:
        datos=maximo.read()
        lista=datos.splitlines()
    for cliente in lista:
        print(cliente)
        
```

```python
precios=[7,16]
print("Grabar pedidos en fichero binario")
print("=================================")
num_col=int(input("Introduzca el número de pedidos del sistema: "))
Leer_y_Grabar_Pedidos("pedidos.dat", num_col)
print()
print("Contenido fichero binario")
print("=========================")
Mostrar_Fichero_Binario("pedidos.dat",precios) 
print()
print("Filtrar por máximo")
print("==================")
importe_maximo=float(input("Introduzca importe máximo: "))
Calcular_Maximos_Clientes("pedidos.dat",importe_maximo, precios)
print()
print("Contenido fichero texto")
print("=======================")
Mostrar_Fichero_Texto()
```

### Genera una función `Translate(text,dicc)` que reciba un texto sencillo y el diccionario generado en el apartado anterior y devuelva un string con la traducción palabra a palabra del texto de entrada. 

En caso de incluir una palabra para la que no existe traducción, la función debe dejar la palabra original pero resaltada con comillas. Por ejemplo:

        Translate("February is the best month. Because February only has 28 days",dicctionary)
        "febrero es el mejor mes. porque febrero solo tiene '28' 'days' "

En caso de no haber realizado el apartado anterior puedes hacer uso de este diccionario:

                     "month":"mes", "because":"porque",
                     "only":"solo","has":"tiene"}

### **Solución 7.2:**

```python
def Translate(text,dicc):
    words=text.split(" ")
    out=""
    for w in words:
        english=w.lower()
        if "." in english: 
            spanish=dicc.get(english.replace(".",""),"'"+english+"'")
            out=out+spanish+". "
        else:
            spanish=dicc.get(english,"'"+english+"'")
            out=out+spanish+" "
    return out
```

```python
def search_words(spanish_words, word):   
    # Obtención de los índices de la palabra que contienen letras
    # ===========================================================
    indices=[]
    for index in range(len(word)):
        if word[index]!="_":
            indices.append(index)
    
    # Recorrer la lista de palabras buscando aquellas palabras 
    # que hacen match con esas letras
    # ========================================================
    candidatos=[]
    for palabra in spanish_words:
        match=True
        if len(palabra)!=len(word):
            match=False
        else:
            for i in indices:
                if word[i]!=palabra[i]:
                    match=False
                    break
        if match==True:
            candidatos.append(palabra)
        
    return candidatos
```
